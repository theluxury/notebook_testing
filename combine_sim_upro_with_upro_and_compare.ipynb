```python
import pandas as pd

simmulated_upro = pd.read_csv('/Users/zhengkanwang/Documents/programming/notebook_testing/historical_data/UPROSIM_WITH_NEW_DATES.csv')
actual_upro = pd.read_csv('/Users/zhengkanwang/Documents/programming/notebook_testing/historical_data/UPRO.csv')
```

```python
simmulated_upro.set_index('Date', inplace=True)
actual_upro.set_index('Date', inplace=True)
```

```python
joined_df = simmulated_upro.join(actual_upro, how='inner')
```

```python
joined_df
```

```python
joined_df['adjusted_upro_sim'] = joined_df['URPO_PRICES'] / 1.213504
joined_df['adjusted_upro_actual'] = joined_df['Close'] / 1.205556
```

```python
joined_df
```

```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
output_notebook()

joined_df.to_csv('/Users/zhengkanwang/Documents/programming/notebook_testing/historical_data/joined_pro_test.csv')
```

```python
import numpy as np

p = figure(title="Simple line example", y_axis_label='price', x_axis_type="datetime")
p.line(np.array(joined_df.index, dtype=np.datetime64),joined_df['adjusted_upro_sim'], legend_label="Simulated.", line_width=2, color='red')
p.line(np.array(joined_df.index, dtype=np.datetime64), joined_df['adjusted_upro_actual'], legend_label="Actual", line_width=2, color='blue')
show(p)
```

```python
p = figure(title="Simple line example", y_axis_label='price', x_axis_type="datetime")
p.line(np.array(joined_df.index, dtype=np.datetime64),joined_df['URPO_PRICES'] / joined_df['Close'], legend_label="Simulated.", line_width=2, color='red')
show(p)
```
